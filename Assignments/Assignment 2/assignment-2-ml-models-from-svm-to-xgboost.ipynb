```python
import pandas as pd
from scipy.stats import skew

# Separate column types
categorical_cols = df.select_dtypes(include='object').columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Fill missing values in categorical columns with mode
for col in categorical_cols:
    if df[col].isnull().any():
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)

# Fill missing values in numerical columns
for col in numerical_cols:
    if df[col].isnull().any():
        col_skew = skew(df[col].dropna())  # avoid skewing from NaNs
        if abs(col_skew) < 0.5:
            df[col].fillna(df[col].mean(), inplace=True)
        else:
            df[col].fillna(df[col].median(), inplace=True)
            
```

## cross-validation to fine-tune hyperparameters for Support Vector Classifier

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define parameter grid
param_grid = {
    'C': [0.1, 1], # Regularization strength (smaller = more regularization)
    'kernel': ['linear', 'rbf', 'poly'], # 'linear', 'rbf', 'poly', 'sigmoid'
    'gamma': ['scale', 0.1]  # Kernel coefficient for 'rbf', 'poly', and 'sigmoid' kernels 
}

# Initialize GridSearchCV
grid_search = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy')

# Fit on training data
grid_search.fit(x_train, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)

# Evaluate on test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

## cross-validation for random forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold

rfc = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# When class labels are imbalanced, KFold might split them unevenly across folds. StratifiedKFold ensures each fold maintains the same class distribution as the full dataset.
scores = cross_val_score(rfc, x, y, cv=skf, scoring='accuracy')

print("Stratified Cross-Validation Accuracy Scores:", scores)
print("Mean Accuracy:", scores.mean())
```
